```python
# Convert Affects_Academic_Performance from Yes/No to True/False
df['Affects_Academic_Performance'] = df['Affects_Academic_Performance'].map({'Yes': True, 'No': False})
```

```python
df.head()
```

```python
# Createing a new column 'Is_Addicted' based on Addicted_Score threshold
df['Is_Addicted'] = df['Addicted_Score'] >= 7
```

```python
df.head()
```

```python
df.columns
```

```python
# Group by Most_Used_Platform and calculate mean daily usage
platform_usage = df.groupby('Most_Used_Platform')['Avg_Daily_Usage_Hours'].mean().reset_index()

# Display the result
print(platform_usage)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Barplot using seaborn
sns.barplot(x='Most_Used_Platform', y='Avg_Daily_Usage_Hours', data=platform_usage, palette='coolwarm')
plt.title('Average Daily Usage by Platform')
plt.ylabel('Avg Daily Usage (Hours)')
plt.xlabel('Platform')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count values per age group
age_counts = df['Age_Group'].value_counts()

# Pie chart
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Age Group Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.countplot(x='Is_Addicted', data=df, palette='viridis')
plt.title('Addicted vs Non-Addicted Students')
plt.xlabel('Is Addicted')
plt.ylabel('Count')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='Affects_Academic_Performance', data=df, palette='viridis')
plt.title('Affects Academic Performance')
plt.xlabel('Affects Academic Performance (True/False)')
plt.ylabel('Number of Students')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns for correlation analysis
cols = ['Addicted_Score', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Avg_Daily_Usage_Hours']

# Compute the correlation matrix
corr_matrix = df[cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
